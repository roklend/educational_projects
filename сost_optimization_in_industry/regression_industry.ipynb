Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Необходимо построить модель, которая предскажет температуру стали. <br>
Целевой признак - последняя измеренная температура в партии. Целевая метрика - MAE, как наиболее интуитивно понятной для температуры. Значение MAE лучшей модели на тестовой выборке должно быть не более 6,8.

```python
df_arc = pd.read_csv('datasets/data_arc_new.csv')
df_bulk = pd.read_csv('datasets/data_bulk_new.csv')
df_bulk_time = pd.read_csv('datasets/data_bulk_time_new.csv')
df_gas = pd.read_csv('datasets/data_gas_new.csv')
df_temp = pd.read_csv('datasets/data_temp_new.csv')
df_wire = pd.read_csv('datasets/data_wire_new.csv')
df_wire_time = pd.read_csv('datasets/data_wire_time_new.csv')
```

#### df_arc

Эта таблица содержит данные об электродах, которые подогревают расплав.

```python
data = df_arc
num_col = list(data.columns)
for col in data.columns:
    if (data[col].dtype != 'int64') & (data[col].dtype != 'float64'):
        num_col.remove(col)
```

```python
display(data.sample(5));
display(data.info());
display(data.describe());
```

Итак, она содержит 5 столбцов: <br>
1) *key* - номер партии; <br>
2) *Начало нагрева дугой* - время начала операции нагрева; <br>
3) *Конец нагрева дугой* - время конца операции нагрева дугой; <br>
4) *Активная мощность* - активная мощность на подогревных электродах; <br>
5) *Реактивная мощность* - реактивная мощность на подогревных электродах. <br>
Пропусков в данных нет, это хорошо. Проверим данные на дубликаты. <br>

```python
data.duplicated().sum()
```

Взгялнем на то, как распределены численные данные.

```python
position = 0
num_region = None
plt.figure(figsize=[15, 20])
plt.subplots_adjust(top=0.94, wspace=0.5, hspace=0.5)
plt.suptitle('ГРАФИЧЕСКИЙ АНАЛИЗ df_arc', fontsize=15)

for col in num_col:
    position += 1
    plt.subplot(5, 2, position)
    sns.set_style('whitegrid')
    plt.title(f'BOXPLOT ДЛЯ {col.upper()}', fontsize=10)
    sns.set_style('whitegrid')
    sns.boxplot(data=data[col], orient='h')
    plt.axvline(np.mean(data[col]), 
                color='r', 
                linestyle='-',
                label='СРЕДНЕЕ')
    plt.legend(title=f'СРЕДНЕЕ = {np.mean(data[col]):.2f}, МЕДИАНА = {np.median(data[col]):.2f}', 
               loc='center', 
               edgecolor = 'r',
               facecolor = 'oldlace',
               bbox_to_anchor=(0.5, -0.24))
    position += 1
    plt.subplot(5, 2, position)
    sns.set_style('whitegrid')
    plt.title(f'Гистограмма ДЛЯ {col.upper()}', fontsize=10)
    sns.histplot(data[col], color='b', label='Сырье', kde=False, bins=30)
    
plt.show()
```

Со столбцами *key* и *Активная мощность* все в порядке. А в столбце *Реактивная мощность* явная ошибка - реактивная мощность не может быть отрицательной.

```python
# Удаляем партию, где присутствует отрицательная мощность
key_to_delete = df_arc[df_arc['Реактивная мощность'] < 0]['key']
df_arc = df_arc[~df_arc['key'].isin(key_to_delete)]
df_arc.describe()
```

```python
# Создадим новые признаки
df_arc['Начало нагрева дугой'] = pd.to_datetime( df_arc['Начало нагрева дугой'], format='%Y-%m-%d %H:%M:%S.%f')
df_arc['Конец нагрева дугой'] = pd.to_datetime( df_arc['Конец нагрева дугой'], format='%Y-%m-%d %H:%M:%S.%f')

df_arc['Время нагрева'] = (df_arc['Конец нагрева дугой'] - df_arc['Начало нагрева дугой']).dt.seconds
df_arc['Полезная работа тока'] = df_arc['Время нагрева'] * df_arc['Активная мощность']

df_arc['Циклы нагрева'] = 1

df_arc['Коэффициент мощности'] = df_arc['Активная мощность'] / (df_arc['Активная мощность']**2 + df_arc['Реактивная мощность']**2)**0.5

df_arc['Полная мощность'] = (df_arc['Активная мощность']**2 + df_arc['Реактивная мощность']**2)**0.5
df_arc['Полная работа тока'] = df_arc['Полная мощность'] * df_arc['Время нагрева']
df_arc['Паразитная работа тока'] = df_arc['Реактивная мощность'] * df_arc['Время нагрева']

df_arc
```

```python
df_arc = df_arc.groupby('key').agg(
    {
     'Полезная работа тока':sum,
     'Время нагрева':sum,
     'Циклы нагрева':sum,
     'Начало нагрева дугой':'first',
     'Конец нагрева дугой':'last',
     'Коэффициент мощности' : 'mean',
     'Паразитная работа тока' : 'sum',
     'Полная работа тока' : 'sum'
    }
)

df_arc.sample(5)
```

### Отчет по решению

#### Цель проекта
Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали.<br>
Сталь обрабатывается в ковше, вместимостью 100 тонн (рисунок). Каждый ковш - это партия, имеющая свой индекс *key* в данных. <br>

Технологический процесс обработки стали состоит из следующих этапов: <br>
1) Первичный  замер температуры расплава (до операций обработки); данные температуры и время замера сохраняются в df_temp; <br>
2) Технологический нагрев расплава электродами; информация о нагреве сохраняется в df_arc <br>
3) Введение легирующих добавок; количественная информация по добавкам сохраняется в df_bulk и df_wire; <br>
4) Продувка сплава инертным газом и перемешивание; информация о количестве израсходованного газа на всю партию сохраняется в df_gas; <br>
5) Повторное измерение температуры и химического состава; если состав и температура не являются оптимальными, возврат к пункту номер 2; <br>
<br>
<div class="alert" style="background-color:#ead7f7;color:#8737bf">
Основной потребитель энергии - дуговой разряд, которым нагревают расплав с использованием электродов. Чтобы уменьшить затраты на электроэнергию, нужно обучить модель машинного обучения, способную решить задачу регрессии для предсказания оптимальной температуры расплава. <br>
<br>
Целевой признак - последняя измеренная температура (оптимальная) партии; <br>
Метрика оценк* - MAE. <br> Необходимо достигнуть MAE < 6.8

#### Выбранная модель
Выбранная по итогам сравнения модель - XGB Regressor (eXtreme Gradient Boosting Regressor). Она показала MAE = 5.87 на тестовой выборке, затратив на обучение около 30 секунд. Xgboost — это библиотека машинного обучения, которая реализует алгоритмы повышения градиента ( деревья решений с градиентным усилением ). Деревья решений с градиентным усилением - это тип алгоритма машин с градиентным усилением, который имеет множество деревьев решений в ансамбле. Все эти деревья решений, как правило, являются слабыми предикторами, и их предсказания объединяются для получения окончательного предсказания. <br>
Подобранные в ходе решения проекта гиперпараметры лучшей модели: <br>
- random_state = 190623. random_state используется в качестве начального числа для генератора псевдослучайных чисел в scikit-learn, чтобы дублировать поведение, когда такая случайность задействована в алгоритмах. Когда зафиксировано random_state, оно будет давать одинаковые результаты при разных запусках программы; <br>
- n_estimators = 1000. Количество деревьев с градиентным усилением; <br>
- max_depth = None. Максимальная глубина деревьев. None означает, что модель неплохо работает из "коробки", сама определяя оптимальное значение глубины; <br>
- learning_rate = 0.01. Коэффициент скорости обучения – это гиперпараметр, определяющий порядок того, как мы будем корректировать наши весы с учётом функции потерь в градиентном спуске. Также можно назвать его величиной шага, с которым мы двигаемся к минимуму функции потерь. Больше шаг = больше погрешность. Слишком маленький шанс - возможное переобучение модели.
